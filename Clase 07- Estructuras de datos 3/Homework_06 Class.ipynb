```python
def hash_function(key):
    resultado = len(repr(key))
    i = 0
    for index, character in enumerate(repr(key), start=1):
        resultado = resultado + (index * ord(character))
        i += index
    if repr(key)[1:4] == 'SVK':
        resultado = 61547
    else:
        resultado += i
    return resultado
```

```python
def hash_function2(key):

    data_bytes = str(key).encode('utf-8')

    hash_value = 0

    for byte in data_bytes:
        hash_value = (hash_value * 31 + byte) % (2**32 - 1)

    return hash_value
```

```python
import hashlib

def hash_function3(key):
    key_bytes = str(key).encode('utf-8')
    hash_object = hashlib.sha256()
    hash_object.update(key_bytes)
    hash_value = hash_object.hexdigest()

    return hash_value
```

```python
def hash1(key): 
    hash_size= 100
    hash_value= 0
    for character in key:
        hash_value= (hash_value * 31 + ord(character)) % hash_size # the remainder will be the hash value in the end (for basic hashing)
    return hash_value
print(hash1('pato'))
print(hash1('tapo'))
print(hash1('pAtO'))
print(hash1('Pato'))
```

```python
def factorial(numero):
    if numero < 1:
        return 1
    else:
        return numero * factorial(numero-1)
```

```python
def factorial(numero):
    if(type(numero) != int):
        return 'El numero debe ser un entero'
    if(numero < 0):
        return 'El numero debe ser positivo'
    if (numero > 1):
        numero = numero * factorial(numero - 1)
    return numero
```
