Consideraciones: Se puede utilizar la función provista.


```python
def hash_function(key):
    return sum(index * ord(character) for index, character in enumerate(repr(key), start=1)) 
```

```python
print(hash_function('pato'))
print(hash_function('tapo'))
```

```python
def hash1(key): 
    hash_size= 100
    hash_value= 0
    for character in key:
        hash_value= (hash_value * 31 + ord(character)) % hash_size # the remainder will be the hash value in the end (for basic hashing)
    return hash_value
```

```python
hash1('@')
```

```python
def hash2(key): #takes a string
    hash_size= 100
    hash_value= 0
    mult = 31 # A prime number improves distribution in hashing
    for character in key: # We iterate over all character of our key
        hash_value= ((hash_value * mult) + ord(character)) % hash_size # math to retrieve the remainder which will be the hash value in the end
    return hash_value                    # ord() will add the ASCII code for that particular character.
```

```python
hash2('@')
```
